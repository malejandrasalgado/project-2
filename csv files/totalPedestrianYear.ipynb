```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

file_to_load="C:/Users/aleja/OneDrive/Bootcamp/projects/project-2/csv files/peds_data.csv"
data = pd.read_csv(file_to_load)
data
```

```python
New_df = data[['data_date','dailyTotal']].copy()
New_df
```

```python
daily_peds_daily=New_df.groupby(['data_date',])['dailyTotal'].sum()
daily_data = daily_peds_daily.reset_index()
daily_data['data_year'] = daily_data['data_date'].str.slice(-4)
daily_data["2019"]= np.nan
daily_data["2020"]= np.nan
daily_data["data_day"]= daily_data["data_date"].str.slice(0,2)
for index, row in daily_data.iterrows():
    if row["data_year"] == "2019":
        val2019 = row["dailyTotal"]
    else:
        daily_data.loc[index, "2019"]= val2019
        daily_data.loc[index, "2020"]= row["dailyTotal"]
        
    day = row["data_day"]

    if day[1] == "/":

            daily_data.loc[index, "data_day"]= "0"+ day[0] 

final = daily_data.query('data_year == "2020"')
final = final[['data_day','2019','2020']].copy()
final = final.sort_values(by ='data_day' )
final.set_index('data_day',inplace=True)
final.to_csv('C:/Users/aleja/OneDrive/Bootcamp/projects/project-2/csv files/DailyTotsByyear.csv')
```
