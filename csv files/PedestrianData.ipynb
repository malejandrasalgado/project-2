```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# File name strings used to make up to name of the files
inputPath = 'C:/Users/aleja/OneDrive/Bootcamp/projects/Second project/csv files/'
inputpartA = "COM_24PM_"
inputpartB = "-08-20"
inputpartB2 = ""
inputpartC = ".csv"
inputDay = ""
# List for data frames to hold the input from each csv as it is read
file_dfs = []

# This bit is the loop to create the name of the input file to load and then load a df by reading the CSV
for x in range(0,2):
    
    if x==0:
        inputpartB2 = "19"
    else:
        inputpartB2 = "20"
        
    for i in range(1,32):
        if i < 10:
            inputDay = "0"+str(i)
        else:
            inputDay = str(i)
        input_file = inputPath+inputpartA+inputDay+inputpartB+inputpartB2+inputpartC
        #print(input_file)
        #Append it to the master df
        try:
                print("Append ", input_file)
                csv_df = pd.read_csv (input_file,skiprows=([i for i in range(0, 8)] + [j for j in range(86, 248)]))
                col_list= list(csv_df)
                csv_df["dailyTotal"]= csv_df[col_list].sum(axis=1)
#                 col_list= list(csv_df)
#                 col_list.remove('dailyTotal')
                csv_df['avgDaily'] = csv_df[col_list].mean(axis=1)
                csv_df["data_date"]= inputDay+inputpartB+inputpartB2
                file_dfs.append(csv_df)
        except:
                print(input_file," not found")
                continue

            
# concat all of the input files into one master dataframe
master_df = pd.concat(file_dfs)
master_df.to_csv('peds_data.csv')
```
